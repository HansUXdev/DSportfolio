## Kelly Criterion and Risk Management

The Kelly Criterion is a mathematical formula used to determine the optimal size of a series of bets or investments. In this analysis, we use the Kelly Criterion to guide position sizing for each strategy component. However, it's important to note that we only use 1/3 of the calculated Kelly Fraction for several reasons:

1. Risk Mitigation: The full Kelly Criterion can lead to aggressive position sizing, which may be too risky for most investors. Using 1/3 of the Kelly Fraction significantly reduces the risk of large drawdowns.

2. Model Uncertainty: Our analysis is based on historical data and assumptions about future market behavior. By using a fraction of the Kelly Criterion, we account for potential inaccuracies in our model or unexpected market events.

3. Capital Preservation: A more conservative approach helps preserve capital over the long term, which is crucial for the core objective of accumulating shares in SPY and QQQ.

4. Volatility Smoothing: Using a reduced Kelly Fraction helps smooth out the equity curve, potentially leading to a more psychologically manageable investment experience.

By combining these strategic elements - optimized DCA timing, options for hedging and income, and conservative position sizing based on a fractional Kelly Criterion - this approach aims to provide a sophisticated framework for long-term investors seeking to build significant positions in broad market ETFs while managing risk and potentially enhancing returns.
