
# Exploratory Data Analysis (EDA) on Financial Data

In this notebook, we will perform exploratory data analysis (EDA) on financial data using various utility functions defined in the `organized_clean_code.py` script. The objectives are to:
1. Fetch historical price data for a specified ticker.
2. Visualize the price data.
3. Calculate and analyze technical indicators.
4. Fetch and analyze fundamental data.
5. Explore options data.

## Libraries and Setup


```python

# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
# Utility functions (imported from the given script)
from data_collection.data_processing import (
    get_fundamentals_data, fetch_fundamentals,
    bollinger_bands, macd, rsi, woodie_pivots, obv, atr,
    get_options_data, get_spy_options, 
    load_price_data, fetch_financial_data
)
# Adjust script_name with correct module or import method as needed
```


## Fetching Historical Price Data

We will start by fetching the historical price data for the ticker 'AAPL' (Apple Inc.) from Yahoo Finance.


```python

# Fetch historical price data for Apple Inc.
ticker = 'AAPL'
start_date = '2010-01-01'
end_date = '2023-01-01'

# Using the load_price_data function from the script
price_data = load_price_data(ticker, start=start_date, end=end_date)
price_data.head()
```


## Visualizing Price Data

Let's visualize the historical closing prices for Apple Inc.


```python

# Plot historical closing prices
plt.figure(figsize=(14, 7))
plt.plot(price_data.index, price_data['Close'], label='Close Price')
plt.title('Historical Closing Prices of AAPL')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```


## Calculating Technical Indicators

Now, let's calculate and visualize some commonly used technical indicators such as Bollinger Bands, MACD, and RSI.


```python

# Calculate Bollinger Bands
bollinger_bands(price_data)

# Plot Bollinger Bands
plt.figure(figsize=(14, 7))
plt.plot(price_data.index, price_data['Close'], label='Close Price')
plt.plot(price_data.index, price_data['Bollinger_High'], label='Bollinger High', linestyle='--')
plt.plot(price_data.index, price_data['Bollinger_Low'], label='Bollinger Low', linestyle='--')
plt.fill_between(price_data.index, price_data['Bollinger_High'], price_data['Bollinger_Low'], color='lightgrey', alpha=0.4)
plt.title('Bollinger Bands for AAPL')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate MACD
macd(price_data)

# Plot MACD
plt.figure(figsize=(14, 7))
plt.plot(price_data.index, price_data['MACD'], label='MACD')
plt.plot(price_data.index, price_data['Signal'], label='Signal', linestyle='--')
plt.title('MACD for AAPL')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Calculate RSI
rsi(price_data)

# Plot RSI
plt.figure(figsize=(14, 7))
plt.plot(price_data.index, price_data['RSI'], label='RSI')
plt.title('RSI for AAPL')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
